# Data Inspection

```python
import pandas as pd
import glob
import glob, re

DATA_DIR  = '../data'

SYMBOLS = ['WIFUSDT', 'ZAMAUSDT']
files = glob.glob(f'{DATA_DIR}/{SYMBOLS[0]}_orderbook_*.parquet')
DATES = sorted(re.search(r'(\d{4}-\d{2}-\d{2})', f).group(1) for f in files)
print(f'Detected dates: {DATES}')
```

## 1. Trades data inspection

```python
for symbol in SYMBOLS:
    print(f'\n{'='*50}')
    print(f'  {symbol} — TRADES')
    print(f'{'='*50}')
    frames = []
    for date in DATES:
        path = f'{DATA_DIR}/{symbol}_trades_{date}.parquet'
        df = pd.read_parquet(path)
        frames.append(df)
        print(f'\n[{date}]  shape: {df.shape}')

    df_all = pd.concat(frames)
    print(f'\n--- dtypes ---')
    print(df_all.dtypes)
    print(f'\n--- head(3) ---')
    print(df_all.head(3))
    print(f'\n--- null counts ---')
    print(df_all.isnull().sum())
    print(f'\n--- describe ---')
    print(df_all.describe())
```

## 2. Orderbook data inspection

```python
for symbol in SYMBOLS:
    print(f'\n{'='*50}')
    print(f'  {symbol} — ORDERBOOK')
    print(f'{'='*50}')
    frames = []
    for date in DATES:
        path = f'{DATA_DIR}/{symbol}_orderbook_{date}.parquet'
        df = pd.read_parquet(path)
        frames.append(df)
        print(f'\n[{date}]  shape: {df.shape}')

    df_all = pd.concat(frames)
    print(f'\n--- dtypes ---')
    print(df_all.dtypes)
    print(f'\n--- head(3) ---')
    print(df_all.head(3))
    print(f'\n--- null counts ---')
    print(df_all.isnull().sum())
    print(f'\n--- describe ---')
    print(df_all.describe())
```

## 3. Timestamp index check

```python
# check trade timestamp columns
df_trade = pd.read_parquet(f'{DATA_DIR}/WIFUSDT_trades_2026-04-12.parquet')
df_ob    = pd.read_parquet(f'{DATA_DIR}/WIFUSDT_orderbook_2026-04-12.parquet')

print('--- trades columns ---')
print(df_trade.columns.tolist())
print('\n--- orderbook columns ---')
print(df_ob.columns.tolist())

# Identify and display timestamp-like columns
for col in df_trade.columns:
    if df_trade[col].dtype in ['int64', 'float64', 'datetime64[ns]'] or 'time' in col.lower():
        print(f'\ntrades [{col}]: dtype={df_trade[col].dtype}, min={df_trade[col].min()}, max={df_trade[col].max()}')

for col in df_ob.columns:
    if df_ob[col].dtype in ['int64', 'float64', 'datetime64[ns]'] or 'time' in col.lower():
        print(f'\norderbook [{col}]: dtype={df_ob[col].dtype}, min={df_ob[col].min()}, max={df_ob[col].max()}')
```

## 4. Check aggressor_side flag (trades)

```python
# True = sell aggressor, False = buy aggressor (check against task specification)
for col in df_trade.columns:
    if df_trade[col].dtype == bool or 'side' in col.lower() or 'aggr' in col.lower() or 'buy' in col.lower():
        print(f'[{col}]')
        print(df_trade[col].value_counts())
        print()
```
